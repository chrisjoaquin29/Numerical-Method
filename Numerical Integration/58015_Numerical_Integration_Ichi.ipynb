#**3 DIFFERENT NUMERICAL INTEGRATION FUNCTION USING SCIPY**
##1. QUAD FUNCTION
This function allows the user to compute the definite integrals of an equation. [[1]](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html#scipy.integrate.quad). *f* symbolizes the equation that it needs to be integrated, *a* serves as the lower limit of integration and *b* represents the upper limit of integration. 

```python
###First question 
###Quadpack equation
#Single Integral
f = lambda x: x**2
integrate.quad (f,0,10)
```

##2. DOUBLE QUAD FUNCTION
One of the quad pack packages that can solve double integral. [[2]](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.dblquad.html#scipy.integrate.dblquad) In *f*, it must have at least two variables (x and y), x serves as the first agrument and y is the second argument. The *a* and *b* are the limits of integration.

```python
###Double Integral
f = lambda y, x: x*y**2
integrate.dblquad(f, 0, 2, lambda x: 0, lambda x: 1)
```

##3. ROMBERG INTEGRATION FUNCTION
This function is under of scipy integration function which the romberg integration using samples of a function. [[3]](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.romb.html#scipy.integrate.romb)

```python
###Romberg Integration
x = np.arange(10, 14.25, 0.25)
y = np.arange(3, 12)

y = np.sin(np.power(x, 2.5))
integrate.romb(y)
```

##POLYNOMIAL EQUATIONS
###*ScipyFunction*

```python
###ScipyFunc
###Polynomial Equation 1
def f(x):
  return 3.0*x**4+2*x**2

i,err = integrate.quad(f,-1,1) 
print(i)
print(err)
```

```python
###ScipyFunc
###Polynomial Equation 2
def f(x):
  return x**10-6*x**5+2*x**3

i,err = integrate.quad(f,-2.3,2.3) 
print(i)
print(err)
```

###*Trapeziodal Rule*

```python
###Trapeziodal Rule
###Polynomial Equation 1
def trapz_rule(func,lb,ub,size):
  h = (ub-lb)/size
  return h*(0.5*(func(lb)+func(ub))+np.sum(func(lb+h*np.arange(1,size))))

f = lambda x: 3.0*x**4+2*x**2
sum = trapz_rule(f, -1,1,1e4)
print(sum)
f_s = 2.5333 #calculated integration
err = abs(f_s-sum)
print(err)
```

```python
###Trapeziodal Rule
###Polynomial Equation 2
f = lambda x: x**10-6*x**5+2*x**3
sum = trapz_rule(f, -2.3,2.3,1e4)
print(sum)
f_s = 1732.38 #calculated integration
err = abs(f_s-sum)
print(err)
```

###*Simpson's 1/3 Rule*

```python
###Simpson's 1/3 Rule
######Polynomial Equation 1
def simp_13(func,lb,ub,divs):
  h = (ub-lb)/divs
  A = (func(lb)+func(ub))+ \
            np.sum(4*func(lb+h*np.arange(1,divs,2)))+ \
            np.sum(2*func(lb+h*np.arange(2,divs,2)))
  S = (h/3)*A
  return S

h = lambda x: 3.0*x**4+2*x**2
sum = simp_13(h, -1,1,1e4)
print(sum)
h_s = 2.5333
err = abs(h_s-sum)
print(err)
```

```python
###Simpson's 1/3 Rule
######Polynomial Equation 2
def simp_13(func,lb,ub,divs):
  h = (ub-lb)/divs
  A = (func(lb)+func(ub))+ \
            np.sum(4*func(lb+h*np.arange(1,divs,2)))+ \
            np.sum(2*func(lb+h*np.arange(2,divs,2)))
  S = (h/3)*A
  return S

h = lambda x: x**10-6*x**5+2*x**3
sum = simp_13(h, -2.3,2.3,1e4)
print(sum)
h_s = 1732.38
err = abs(h_s-sum)
print(err)
```

###*Simpson's 3/8 Rule*

```python
###Simpson's 3/8 Rule
######Polynomial Equation 1
def simp_38(func,lb,ub,divs):
  h = (ub-lb)/divs
  A = (func(lb)+func(ub))+ \
            np.sum(3*(func(lb+h*np.arange(1,divs,3))))+ \
            np.sum(3*(func(lb+h*np.arange(2,divs,3))))+ \
            np.sum(2*func(lb+h*np.arange(3,divs,3)))
  S = (3*h/8)*A
  return S

z = lambda x:3.0*x**4+2*x**2
sum = simp_38(z, -1,1,1e4)
print(sum)
z_s = 2.5333
err = abs(z_s-sum)
print(err)
```

```python
###Simpson's 3/8 Rule
######Polynomial Equation 2
g = lambda x: x**10-6*x**5+2*x**3
sum = simp_38(g,-2.3,2.3,1e4)
print(sum)
g_s = 1732.38
err = abs(g_s-sum)
print(err)
```

###*Monte Carlo Rule*

```python
###Monte Carlo Rule
######Polynomial Equation 1
a, b = -1, 1
n = 1e3
samples = np.random.uniform(a,b,int(n))
o = lambda x: 3.0*x**4+2*x**2
A = np.sum(o(samples))
S = ((b-a)/n)
S ###Error
```
